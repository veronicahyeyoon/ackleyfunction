```python
# plotting example graph
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
x,y=np.meshgrid(x,y)
a= -0.2*(((x**2+y**2)/2)**0.5)
b= (np.cos(2*np.pi*x)+np.cos(2*np.pi*y))/2
z= -20*np.exp(a)-np.exp(b)+20+np.e
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.plot_surface(x,y,z, color='pink', alpha=0.5)
ax.scatter(x,y,z, color='purple', alpha=0.003)
plt.xlabel('x')
plt.ylabel('y')

plt.show=()
```

```python
#for loop finding optimal minimal point

points= []

for iteration in range (0,50):

    for test in range(0,5):
        add_to_x= np.random.uniform(low=-1,high=1)
        add_to_y= np.random.uniform(low=-1,high=1)
        new_list= np.array([add_to_x, add_to_y])
        print(f"New small x and y:{new_list}")

        x= np.random.uniform(low=-10,high=10)
        y= np.random.uniform(low=-10,high=10)
        new_x= x+add_to_x
        new_y= y+add_to_y
        new_list_add= np.array([new_x,new_y])
        print(f"New added values of x and y: {new_list_add}")

        z_value= ackleyfunctioncalculator.ackleyinput(new_x,new_y)
        entire_array= np.array([new_x,new_y,z_value])
        print(f"Final xyz array:{entire_array}")

        points.append(entire_array)
        total_matrix= np.array(points)
        print("The line below is the total matrix :)")
        print(f"{total_matrix}")

        find_z_min= np.min(total_matrix[:, 2])
        min_array= np.argmin(total_matrix[:, 2])
        print(f"This is the array with min z value: {min_array}")
        print(f"This is z_min for loop: {find_z_min}")
        final_min_array= total_matrix[min_array]
        print(f"This is the FINAL min z array: {final_min_array}")
        x_y_originalish= np.array(final_min_array)
        total_matrix= x_y_originalish
        print(f"New total matrix: {total_matrix}")

    print("kind of done :/")

    x= np.linspace(-10,10,100)
    y= np.linspace(-10,10,100)
    x,y= np.meshgrid(x,y)
    a= -0.2*(((x**2+y**2)/2)**0.5)
    b= (np.cos(2*np.pi*x)+np.cos(2*np.pi*y))/2
    z= -20*np.exp(a)-np.exp(b)+20+np.e
    fig=plt.figure()
    ax=fig.add_subplot(111,projection='3d')
    ax.plot_surface(x,y,z, color='pink', alpha=0.5)
    ax.scatter(x,y,z, color='purple', alpha=0.003)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.scatter(total_matrix[0],total_matrix[1],total_matrix[2], color='black')
    plt.show=()
    plt.savefig(f"/Users/veronicajung/codefolder/ackleyfunction/graphimages/frame_{iteration}.png")

print("finally done \^0^/")
```
